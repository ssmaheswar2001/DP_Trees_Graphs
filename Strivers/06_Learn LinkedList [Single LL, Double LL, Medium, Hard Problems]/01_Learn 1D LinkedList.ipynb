# Deleting a node in LinkedList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class SingleLL:
    def __init_(self):
        self.head = None
        
    def convertArrToLL(self, arr):
        n = len(arr)
        self.head = Node(arr[0])
        current = self.head
        for i in range(1, n):
            newNode = Node(arr[i])
            current.next = newNode
            current = newNode
        return self.head
    
    def printLL(self, head):
        current = head
        
        while(current != None):
            print(current.data, end='->')
            current = current.next
        print("")
            
    def deleteAtBeginning(self, head):
        if head == None:
            return None
        
        temp = head
        head = head.next
        temp.next = None
        
        return head
    
    def deleteAtPosition(self, head, position):
        if head is None:
            return head
        
        temp = head
        prev = None
        
        if position == 1:
            head = head.next
            temp.next = None
            return head
        
        for i in range(1, position):
            prev = temp
            temp = temp.next
            if temp is None:
                print('Data not found')
                return head
        
        prev.next = temp.next
        temp.next = None
        
        return head
    
    def deleteByValue(self, head, key):
        if head == None:
            return None
        
        if head.data == key:
            temp = head
            head = head.next
            temp = None
            return head
        
        temp = head
        prev = None
        
        while temp.data != key:
            prev = temp
            temp = temp.next
            if temp is None:
                print("Not Found")
                return head
            
        prev.next = temp.next
        temp.next = None
        return head
        
    
    def deleteAtEnd(self, head):
        if head == None or head.next == None :
            return None
        
        temp = head
        prev = None
        
        while temp.next != None:
            prev = temp
            temp = temp.next
            
        prev.next = None
        return head
        

arr = [1, 2, 3, 4, 5, 6, 7, 8]
ll = SingleLL()
head = ll.convertArrToLL(arr)
ll.printLL(head)
head = ll.deleteAtBeginning(head)
ll.printLL(head)
head = ll.deleteAtPosition(head, 3)
ll.printLL(head)
head = ll.deleteByValue(head, 6)
ll.printLL(head)
head = ll.deleteAtEnd(head)
ll.printLL(head)
```

# Inserting a Node in LinkedList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class SingleLL:
    def __init__(self):
        self.head = None
        
    def insertAtBeginning(self, value):
        newNode = Node(value)
        newNode.next = self.head
        self.head = newNode
        return self.head
    
    def insertAtEnd(self, value):
        newNode = Node(value)
        
        if self.head == None:
            self.head = newNode
            return self.head
        
        current = self.head
        while current.next != None:
            current = current.next
            
        current.next = newNode
        return self.head
        
    def insertAtKPosition(self, value, k):
        newNode = Node(value)
        if k == 1:
            newNode.next = self.head
            self.head = newNode
            return self.head
        
        temp = self.head
        for i in range(2, k):
            temp = temp.next
        newNode.next = temp.next
        temp.next = newNode
        
        return self.head        
    
    def printLL(self, head):
        if head == None:
            return None
        
        current = head
        while current != None:
            print(current.data, end='->')
            current = current.next
        print()
        
ll = SingleLL()
head = ll.insertAtBeginning(3)
ll.printLL(head)
head = ll.insertAtBeginning(1)
ll.printLL(head)
head = ll.insertAtEnd(5)
ll.printLL(head)
head = ll.insertAtKPosition(4, 1)
ll.printLL(head)
head = ll.insertAtKPosition(3, 5)
ll.printLL(head)
```

# Length of the LinkedList

```python
# Time Complexity : O(N)
# Space Complexity : O(N)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class singleLL:
    def __init_(self):
        self.head = None
        
    def constructLL(self, arr):
        n = len(arr)

        ll = singleLL()
        ll.head = Node(arr[0])
        current = ll.head

        for i in range(1, n):
            current.next = Node(arr[i])
            current = current.next

        return ll.head

    def printLinkedList(self, head):
        current = head

        while current != None:
            print(current.data, end="->")
            current = current.next
        print()
            
    def Length(self, head):
        current = head
        count = 0
        while current != None:
            current = current.next
            count += 1
        return count
        
            
arr = [1, 2, 3, 4, 5, 6]
ll = singleLL()
head = ll.constructLL(arr)
ll.printLinkedList(head)
print(ll.Length(head))
```

# Search an Element in the LL

```python
# Time Complexity : O(N)
# Space Complexity : O(N)

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class singleLL:
    def __init_(self):
        self.head = None
        
    def constructLL(self, arr):
        n = len(arr)

        ll = singleLL()
        ll.head = Node(arr[0])
        current = ll.head

        for i in range(1, n):
            current.next = Node(arr[i])
            current = current.next

        return ll.head

    def printLinkedList(self, head):
        current = head

        while current != None:
            print(current.data, end="->")
            current = current.next
        print()
            
    def searchLL(self, head, key):
        current = head
        
        while current != None:
            if current.data == key:
                return True
            current = current.next
            
        return False
        
            
arr = [1, 2, 3, 4, 5, 6]
ll = singleLL()
head = ll.constructLL(arr)
ll.printLinkedList(head)
print(ll.searchLL(head, 0))
```
