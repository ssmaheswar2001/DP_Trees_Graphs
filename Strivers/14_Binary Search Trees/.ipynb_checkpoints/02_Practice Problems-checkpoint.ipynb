# Ceil in a Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
        
def findCeil(root, key):
    current = root
    ceil = -1
    while current != None:
        if current.data >= key:
            ceil = current.data
            current =  current.left
        else:
                current = current.right
    return ceil

root = Node(10)
root.left = Node(5)
root.right = Node(11)
root.left.left = Node(4)
root.left.right = Node(7)
root.right.right = Node(12)
findCeil(root, 6)
```

# Floor in BST


```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
        
def findFloor(root, key):
    floor = -1
    current = root
    while current != None:
        if current.data <= key:
            floor = current.data
            current = current.right
        else:
            current = current.left
                
    return floor


root = Node(10)
root.left = Node(5)
root.right = Node(11)
root.left.left = Node(4)
root.left.right = Node(7)
root.right.right = Node(12)
findFloor(root, 6)
```

# Find Kth smallest element in BST

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def kthSmallest(root, k):
    if root is None:
        return None
        
    stack = []
    node = root
    count = 0
    while True:
        if node != None:
            stack.append(node)
            node = node.left
        else:
            if len(stack) == 0:
                break
            node = stack.pop()
            count += 1
            if count == k:
                return node.data
            node = node.right

root = Node(4)
root.left = Node(2)
root.right = Node(7)
root.left.left = Node(1)
root.left.right = Node(3)
kthSmallest(root, 2)
```

# Check if a tree is a BST 

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def isValidBST(root):
    if root is None:
        return True
    prev = -2**31 - 1 
    stack = []
    node = root

    while True:
        if node != None:
            stack.append(node)
            node = node.left
        else:
            if len(stack) == 0:
                break
            node = stack.pop()
            if prev >= node.data:
                return False
            prev = node.data
            node = node.right

    return True

root = Node(4)
root.left = Node(2)
root.right = Node(7)
root.left.left = Node(1)
root.left.right = Node(3)
isValidBST(root)
```

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def isValidBST(node, left, right):
    if node is None:
        return True
        
    isLimit = False
    if left < node.data < right:
        isLimit = True
        
    return isLimit and isValidBST(node.left, left, node.data) and isValidBST(node.right, node.data, right)


root = Node(4)
root.left = Node(2)
root.right = Node(7)
root.left.left = Node(1)
root.left.right = Node(3)
isValidBST(root, -2**31-1, 2**31)
```

# Lowest Common Ancestor of a Binary Search Tree

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def lowestCommonAncestor(root, p, q):
    if root is None:
        return None
    if root == p or root == q:
        return root
    leftNode = lowestCommonAncestor(root.left, p, q)
    rightNode = lowestCommonAncestor(root.right, p, q)
    if leftNode != None and rightNode != None:
        return root
    if leftNode != None:
        return leftNode
    if rightNode != None:
        return rightNode
    return None

root = Node(6)
root.left = Node(2)
root.right = Node(8)
root.left.left = Node(0)
root.left.right = Node(4)
root.left.right.left = Node(3)
root.left.right.right = Node(5)
root.right.left = Node(7)
root.right.right = Node(9)
p = root.left 
q = root.right.right
lowestCommonAncestor(root, p, q).data
```

# Construct Binary Search Tree from Preorder Traversal

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def bstFromPreorder(preorder):
    n = len(preorder)
    if n == 0:
        return None
        
    root = Node(preorder[0])
    i = 1
    while i < n and preorder[i] < preorder[0]:
        i += 1
        
    root.left = bstFromPreorder(preorder[1:i])
    root.right = bstFromPreorder(preorder[i:n])

    return root

preorder = [4, 2, 1, 3, 7, 5, 6]
bstFromPreorder(preorder).data
```

# Inorder Successor/Predecessor in BST
- 

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None


def predecessor(root, key):
    pre = None
    current = root
    while current != None:
        if current.data < key:
            pre = current
            current = current.right
        elif current.data >= key:
            current = current.left
    return pre.data

def successor(root, key):
    suc = None
    current = root
    while current != None:
        if current.data > key:
            suc = current
            current = current.left
        elif current.data <= key:
            current = current.right
    return suc
    
root = Node(5)
root.left = Node(3)
root.left.left = Node(2)
root.left.right = Node(4)
root.left.left.left = Node(1)
root.right = Node(7)
root.right.left = Node(6)
root.right.right = Node(9)
root.right.right.left = Node(8)
root.right.right.right = Node(10)
predecessor(root, 8)
successor(root, 10)
```

# Two Sum In BST | Check if there exists a pair with Sum K

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def findTarget(root, k):
    inorder = []
    stack = []
    node = root
    while True:
        if node != None:
            stack.append(node)
            node = node.left
        else:
            if len(stack) == 0:
                break
            node = stack.pop()
            inorder.append(node.data)
            node = node.right
    i = 0
    j = len(inorder)-1
    while i < j:
        if k == inorder[i]+inorder[j]:
            return True 
        if k > inorder[i] + inorder[j]:
            i += 1
        else:
            j -= 1
    return False

root = Node(5)
root.left = Node(3)
root.left.left = Node(2)
root.left.right = Node(4)
root.right = Node(7)
root.right.left = Node(6)
root.right.right = Node(9)
findTarget(root, 9)
```

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def findTarget(root, k, seen):
    if root is None:
        return False
        
    if (k-root.data) in seen:
        return True
    seen.append(root.data)
    return findTarget(root.left, k, seen) or findTarget(root.right, k, seen)

root = Node(5)
root.left = Node(3)
root.left.left = Node(2)
root.left.right = Node(4)
root.right = Node(7)
root.right.left = Node(6)
root.right.right = Node(9)
seen = []
findTarget(root, 9, seen)
```

# Largest BST in Binary Tree
- 

```python
class Node:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None

def largestBst(root):
    if root is None:
        return (0, 10**5+1, 0)
        
    l, lmin, lmax = largestBst(root.left)
    r, rmin, rmax = largestBst(root.right)
        
    if (lmax < root.data) and (root.data < rmin):
        return (1+l+r, min(root.data, lmin), max(root.data, rmax))
        
    return (max(l, r),0, 10**5+1)
        
root = Node(5)
root.left = Node(2)
root.left.left = Node(1)
root.left.right = Node(3)
root.right = Node(4)
largestBst(root)[0]
```
