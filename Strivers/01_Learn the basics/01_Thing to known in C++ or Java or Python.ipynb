```python
# IF ELSE STATEMENT
# https://practice.geeksforgeeks.org/problems/java-if-else-decision-making0924/0?category%5B%5D=Java&category%5B%5D=Java&difficulty%5B%5D=-2&page=1&query=category%5B%5DJavadifficulty%5B%5D-2page1category%5B%5DJava
'''Given two integers, n and m. The task is to check the relation between n and m.'''

def compareNM(n, m):
        # code here
    if (n < m):
        return 'lesser'
    if (n > m):
        return 'greater'
    return 'equal'

compareNM(3, 8)
```

```python
# Nth Fibonacci Number
# https://www.codingninjas.com/studio/problems/nth-fibonacci-number_74156?utm_source=striver&utm_medium=website&utm_campaign=a_zcoursetuf
'''
The n-th term of Fibonacci series F(n
'''

def nFibonacci(n):
    
    a = 1
    b = 1
    
    for i in range(n-2):
        temp = a + b
        a = b
        b = temp
        
    return b

nFibonacci(6)
```

```python
# While loop- printTable
# https://www.geeksforgeeks.org/problems/while-loop-printtable-java/1

def printTable(n):
    multiplier = 10
    
    while multiplier > 0 :
        print(multiplier*n, end=" ")
        multiplier -= 1
        
printTable(2)
```

### TIME COMPLEXITY
- Time complexity is never called as time take to run the code; Time Complexity != Time Taken
    
- Time complexity is the rate at which the time taken increases with respect to the input size.
    
- Time complexity is said in terms of Big-O notation.
    
- Time complexity is computed in terms of worst case of senario, avoid constants and avoid lower values.    
- Big-O is called Upper Bond.

### Space Complexity
- Memory space
- Space complexity is said in terms of Big-O notation
- Space complexity = Auxilary Space + InputSpace
- Auxilary Space : Space that we take to solve the problem.
- Input Space : The space that we take to store the input.
